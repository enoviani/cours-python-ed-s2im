```python
import matplotlib.pyplot as plt
import numpy as np
```

Tout d'abord, un plot très simple :

```python
plt.plot([1,3,2,4])
```

Puis, quelques gadgets en plus :

```python
t = np.arange(0.0, 5.2, 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
```

Encore plus de contrôle sur la présentation :

```python
t = np.arange(0.0, 5.2, 0.2)
lines = plt.plot(t, t, 'r-', t, t**2, 'b--')
plt.setp(lines[0], linewidth=2.5,
                   label="linear")
plt.setp(lines[1], linewidth=4.,
                   dashes=[10, 2, 6, 4],
                   label="quadratic")
plt.legend()
plt.title("Two curves")
plt.xlabel("X", color="green")
plt.ylabel("Y", color="green")
```

Combiner plusieurs plots :

```python
t = np.arange(-2.*np.pi, 2.*np.pi, 0.2)

fig = plt.figure()

axes_1 = fig.add_subplot(2, 2, 1)
axes_1.plot(t, np.sin(t))

axes_2 = fig.add_subplot(2, 2, 2)
axes_2.plot(t, np.cos(t))

axes_3 = fig.add_subplot(2, 2, 3)
axes_3.plot(t, np.sinh(t))

axes_4 = fig.add_subplot(2, 2, 4)
axes_4.plot(t, np.cosh(t))
```

On peut utiliser des formules en LaTeX :

```python
t = np.arange(0.0001, 4.*np.pi, 0.2)
plt.plot(t, np.sin(t)/t, 'r-')
plt.xlabel(r"$\beta$", fontsize=16)
plt.ylabel(r"$\frac{\sin(\beta)}{\beta}$", fontsize=16)
```
